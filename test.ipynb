```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df = pd.read_csv("titanic.csv")
df.head(11)
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='black', marker = 'o', linestyle = 'dotted')

# add a title
plt.title("Nominal GDP")

# add a label to the y-axisYears
plt.ylabel("Billions of $")
plt.xlabel("Years")
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
flight = sns.load_dataset('flights') # load flight dataset
print(flight)
```

```python
flight_new = flight.groupby('year').sum(numeric_only=True).reset_index()
# line chart
sns.lineplot(x='year',y='passengers',data=flight_new, color='black', marker='o')
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi","Gaule", "West Side Story"]
num_oscars = [5, 11, 3, 8,15, 10]

# plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [num_oscars]
plt.bar(range(len(movies)), num_oscars)

plt.title("My Favorite Movies")     # add a title
plt.xlabel("Movies")
plt.ylabel("# of Academy Awards")   # label the y-axis

# label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)

plt.show()
```

```python
sns.countplot(data=df, x='Sex',hue='Survived')
plt.xlabel("Sex/Survived")
df['Pclass'].value_counts()
```

```python
sns.barplot(x='Pclass', y='Age', hue='Sex', data=df)
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%0.3f')
```

```python
sns.boxplot(df['Age'])
```

```python
sns.boxplot(data=df, x='Sex', y='Age', hue='Survived')
```

```python
plt.hist(df['Age'])
```

```python
sns.displot(df['Age'], kind='hist', kde='True')
```

```python
# Bivariant analysis
sns.kdeplot(df[df['Survived'] == 0]['Age'], label='Died')
sns.kdeplot(df[df['Survived'] == 1]['Age'], label='Survived')

plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
tips = sns.load_dataset('tips')
print(tips)
```

```python
flight_new = tips.groupby('tip').sum(numeric_only=True).reset_index()
# line chart
sns.lineplot(x='total_bill',y='tip',data=flight_new, color='black', marker='o')
```

```python
sns.scatterplot(x='total_bill', y='tip',hue='sex', data=tips, style='smoker')
plt.show()
```

```python
sns.heatmap(pd.crosstab(df['Pclass'],df['Survived'])) 
```

```python
sns.heatmap(flight.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap(flight.pivot_table(values='passengers',index='month',columns='year'))
```
