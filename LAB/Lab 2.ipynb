```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df = pd.read_csv("Health_heart_experimental.csv")
df.head(11)
```

```python
df_sorted = df.sort_values(by='age', ascending=False)
sns.lineplot(x='age',y='SysBP' ,data=df_sorted, color='black', marker='o')
```

From the above line plot, the normal upper BP is 120mmHg. Looking to the graph, the higher BP has gone upto 126.5 mmHg at the age of 60-63. It means for this age group of people, the BP is generally high, it might be due to the environment and definitely due to the age. They would think much about their kids and even their grand childrens. 
And looking for the lowest upper BP compared in this graph, it is 124.5-125.0 for the age of 30-35. It is due to the either they have to work very hard and for long time a day so they don't get to eat properly, also not a good sleep.
Generally, if we see the graph, 20 - 60 years old people have high BP because it is a age where we have to work alot for ourselves, parents and childs.

```python
age_groups = df.groupby('age')['BMI'].mean().reset_index()
sns.lineplot(x='age', y='BMI', data=age_groups, color='black', marker='o')
```

Here, the normal BMI value is 18.5 - 24.9. Analyzing the graph, the first thing is that the every age group has the high weight, high obesity because the BMI value is higher for every age group. From 20 - 30 age group, the BMI value is highest. This means the new generations are more prone to heart disease and also the mental health issues. The main reason for this BMI value at this age is simple that their eating style is fully unhealthy. Like for the early age, they eat chips, junk foods and many more which has more calories and helps in increasing the obesity. And for 25 - 30 age, they have to go to offices and mostly in snacks time, they prefer eating in resturants or any hotels where the kind of food we get is the same as unhealthy.
And also looking for all ages, they eat the unhealthy foods. And also no age group has the time for their health means no time for exercise, yoga and other things.

```python
# Create age groups
bins = [0, 20, 30, 40, 50, 60, 100]
labels = ['0-20','21-30','31-40','41-50','51-60','61+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Average BP by age group
avg_bp = df.groupby('age_group')[['SysBP','DiaBP']].mean().reset_index()
plt.ylabel("SysBP/DiaBP")
# Plot
sns.barplot(x='age_group', y='SysBP', data=avg_bp, color='skyblue', label='Systolic')
sns.barplot(x='age_group', y='DiaBP', data=avg_bp, color='salmon', label='Diastolic')
# sns.barplot(x='age_group', y='SysBP', hue='DiaBP', data=avg_bp)
```

As already mention above, the every age group that higher BP. But the both uppper and lower BPs are decreasing with the increase of ages. But its shocking that the younger age group has the higher BP. Its due to the obesity, no exercise, no yoga, and also no health concerned food items.
