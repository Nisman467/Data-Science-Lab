##Matrix Operations 

```python
A = np.array([
    [2, -1],
    [0, 3]
])
print(A.shape)
```

```python
B = np.array([
    [1, 2],
    [-3, 4]
])
print(B.shape)
```

```python
print("Adding the two matrices : ")
print(A + B)
```

```python
print("Dot Product: ")
print(A @ B)
```

```python
print("Performing the \"3*A - 2*B\"")
print(3 * A - 2 * B)
```

##Determminants and Inverse

```python
import numpy.linalg as LA
```

```python
print("Determinants of :")
print(A)
print("-"*10)
LA.det(A)
```

Here, the determinant of A is 6. Therefore, it is non-singular matrix and has its inverse.

```python
print("Inverse of the matrix A :")
print(A)
print("-"*30)
print(LA.inv(A))
```

##Eigen Value and Eigen Vectors

```python
C = np.array([
    [4, 1],
    [-2, 1]
])
print(C.shape)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(C)

print("Eigen Values = ",eigenvalues[0])
print("Eigen Vectors : ")
print("First Eigen Vector: ",eigenvectors[:,0])
print("Second Eigen Vector: ",eigenvectors[:,1])
```

##Machine Learning

```python
X = np.array([
    [4, 1.5, 100],
    [6, 2.0, 150]
])

y = np.array([20, 30])
```

```python
beta = np.linalg.pinv(X) @ y
```

β=(X.T * X)^−1  X.T y, we don't use this beacuse normal inverse matrix doesn't exists means the shape of X matrix isnot square and singular matrix too. So the equation is X^-1 .y and we use "pinv" means pesudo-inverse done when the matrix isn't square matrix and not have sufficent values. 

```python
car3 = np.array([8, 3.0, 200])

price = car3 @ beta 
```
